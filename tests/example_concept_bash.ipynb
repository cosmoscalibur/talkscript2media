```python
%%bash


#En este caso todo se genero a partir de 3 imagenes 


#Los archivos .txt nombrados 01.txt,02.txt,03.txt con el texto correspondientes a cada una de las imagenes


#Generando  audio .wav a partir de archivos .txt 

#Generando archivos de texto con Festival 

text2wave Texto/01.txt  -o Audio/01.wav

text2wave Texto/02.txt  -o Audio/02.wav

text2wave Texto/03.txt  -o Audio/03.wav


#Extrayendo duracion en segundos con sox y poblando archivo de Duracion,con imagen correspondiente.

echo "ffconcat version 1.0" >> Duracion

echo "file  IMG/01.jpg" >> Duracion # file 01.jpg 

d1=$(sox --info -D Audio/01.wav)

echo 'duration' $d1>>Duracion 

echo "file  IMG/02.jpg" >> Duracion # file 02.jpg

d2=$(sox --info -D Audio/02.wav)

echo 'duration' $d2>>Duracion

echo "file  IMG/03.jpg" >> Duracion # file 03.jpg

d3=$(sox --info -D Audio/03.wav)

echo 'duration' $d3>>Duracion
 
#cat Duracion


#-------------Generando video a partir de imagenes y archivo duracion--------- 



ffmpeg -i Duracion  -vf "scale=trunc(iw/2)*2:trunc(ih/2)*2"  Video1.mp4



#--------------------Union de Audios con sox---------------------- 

sox Audio/01.wav Audio/02.wav Audio/03.wav Audio/long.wav


#-------------------Union Video con Audio-------------------------

ffmpeg -i Video1.mp4 -i Audio/long.wav  Video.mp4



```
